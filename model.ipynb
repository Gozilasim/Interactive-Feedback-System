```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO


# model_m =  YOLO("E:/UTeM2/FYP1/project/Gesture_recognition/runs/detect/train13/weights/best.pt")
# model_l =  YOLO("Testing_Result/yolov8l/detect/train12/weights/best.pt")  
# model_n =  YOLO("Testing_Result/yolov8n/train10/weights/best.pt")  
# model_s =  YOLO("Testing_Result/yolov8s/train11/weights/best.pt")    

# Load YOLOv8 model
model = YOLO("E:/UTeM2/FYP1/project/Gesture_recognition/runs/detect/train13/weights/best.pt")    # Replace with your YOLOv8 model path

image_predict = "WIN_20240922_21_43_34_Pro.jpg"

# Predict on the image
results = model.predict(image_predict, save=True)

# Load the image using cv2
image_path = image_predict
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for displaying with matplotlib

# Process YOLOv8 results and draw on the image
most_confident_box = None

for result in results:
    boxes = result.boxes
    if boxes:
        # Find the most confident detection
        most_confident_box = max(boxes, key=lambda box: box.conf[0])

if most_confident_box:
    x1, y1, x2, y2 = map(int, most_confident_box.xyxy[0])  # Extract bounding box coordinates
    confidence = most_confident_box.conf[0]  # Extract confidence score
    class_id = most_confident_box.cls[0]  # Extract class id

    # Draw bounding box and label on the image
    label = f'{model.names[int(class_id)]}: {confidence:.2f}'
    cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# Display the image with bounding boxes
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
import cv2
import mediapipe as mp
import torch
from ultralytics import YOLO  # Ensure you have the ultralytics library installed

# Load YOLOv8 model
model =  YOLO("E:/UTeM2/FYP1/project/Gesture_recognition/Testing_Result/yolov8s/train11/weights/best.pt")  # Replace with your YOLOv8 model path

# Initialize Mediapipe holistic model
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils

cap = cv2.VideoCapture(0)

# Initiate holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Process frame with YOLOv8
        yolo_results = model(frame)

        # Print YOLOv8 results
        print("YOLOv8 Results:")
        for result in yolo_results:
            boxes = result.boxes
            for box in boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # Extract bounding box coordinates
                confidence = box.conf[0]  # Extract confidence score
                class_id = box.cls[0]  # Extract class id

                # Print each detected object's details
                print(f'Class: {model.names[int(class_id)]}, Confidence: {confidence:.2f}, Box: ({x1}, {y1}, {x2}, {y2})')

                # Draw bounding box and label on the frame
                label = f'{model.names[int(class_id)]}: {confidence:.2f}'
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        # Recolor feed for holistic model processing
        image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = holistic.process(image_rgb)

        # Print Mediapipe holistic model results
        print("Mediapipe Holistic Results:")
        if results.right_hand_landmarks:
            print("Right hand landmarks detected")
        if results.left_hand_landmarks:
            print("Left hand landmarks detected")

        # Draw holistic model results on the frame
        # Face landmarks

        # Right hand
        if results.right_hand_landmarks:
            mp_drawing.draw_landmarks(frame, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
        # Left hand
        if results.left_hand_landmarks:
            mp_drawing.draw_landmarks(frame, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

        cv2.imshow('Holistic Model and YOLOv8 Detections', frame)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import torch
from ultralytics import YOLO  

# Load YOLOv8 model
model =  YOLO("E:/UTeM2/FYP1/project/Gesture_recognition/runs/detect/train10/weights/best.pt")  

# Initialize Mediapipe holistic model
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils

cap = cv2.VideoCapture(0)


# Initiate holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Process frame with YOLOv8
        yolo_results = model(frame)

        # Print YOLOv8 results
        print("YOLOv8 Results:")
        most_confident_box = None

        for result in yolo_results:
            boxes = result.boxes
            if boxes:
                # Find the most confident detection
                most_confident_box = max(boxes, key=lambda box: box.conf[0])

        if most_confident_box:
            x1, y1, x2, y2 = map(int, most_confident_box.xyxy[0])  # Extract bounding box coordinates
            confidence = most_confident_box.conf[0]  # Extract confidence score
            class_id = most_confident_box.cls[0]  # Extract class id

            # Print the most confident detection's details
            print(f'Class: {model.names[int(class_id)]}, Confidence: {confidence:.2f}, Box: ({x1}, {y1}, {x2}, {y2})')

            # Draw bounding box and label on the frame
            label = f'{model.names[int(class_id)]}: {confidence:.2f}'
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        # Recolor feed for holistic model processing
        image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = holistic.process(image_rgb)



        # Draw holistic model results on the frame
        # Right hand
        if results.right_hand_landmarks:
            mp_drawing.draw_landmarks(frame, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
        # Left hand
        if results.left_hand_landmarks:
            mp_drawing.draw_landmarks(frame, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

        cv2.imshow('Holistic Model and YOLOv8 Detections', frame)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
